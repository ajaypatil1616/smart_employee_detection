```python
import pytesseract
import cv2
from gtts import gTTS
import os
from playsound import playsound
import io
import tempfile

# Pre-defined employee data
employee_data = {
    'DITPL263': 'Ajay Patil',
    'DITPL123': 'John Doe',
    # Add more employee IDs and names as needed
}

def play_audio(message):
    """
    Convert text to speech and play the audio without saving the file.
    """
    # Generate speech object
    speech_obj = gTTS(text=message, lang='en', slow=False)
    
    # Save speech to a BytesIO buffer (in-memory file)
    with io.BytesIO() as audio_buffer:
        speech_obj.save(audio_buffer)
        audio_buffer.seek(0)  # Go to the beginning of the buffer
        playsound(audio_buffer)  # Use playsound to play the audio

# Start video capture
cap = cv2.VideoCapture(0)  # 0 is the default camera index

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame. Exiting...")
        break
    
    # Display the video feed
    cv2.imshow('Employee Detection', frame)
    
    # Perform OCR to detect text
    text_detected = pytesseract.image_to_string(frame)
    print("Detected Text:", text_detected)
    
    # Extract employee ID
    text_detected_list = text_detected.split(' ')
    employee_id = ''
    for word in text_detected_list:
        if word.startswith('DIT'):
            employee_id = word[:8]
            break  # Stop searching after finding an ID
    
    if employee_id in employee_data:
        # Play welcome audio for recognized employee
        name = employee_data[employee_id]
        text_to_speech = f"Welcome {name}"
        print(text_to_speech)
        play_audio(text_to_speech)
    elif employee_id:
        # Invalid or unrecognized ID
        text_to_speech = "Please try again"
        print(text_to_speech)
        play_audio(text_to_speech)
    
    # Exit condition
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Exiting...")
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
